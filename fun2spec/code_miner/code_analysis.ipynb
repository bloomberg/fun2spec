```python
from code_miner import CodeMiner
source = '../analysis_code/bde_public'

miner = CodeMiner(source=source, use_cache=False)
```

```python
from collections import Counter
cpp_functions = miner.fns
print('Total number of functions:', len(cpp_functions))
type_count = Counter([cpp_function.return_type.category for cpp_function in cpp_functions])
print('Return type distribution of these functions:', type_count)

print('Total number of functions with docstring:', len([cpp_function for cpp_function in cpp_functions if len(cpp_function.docstring) > 0]))

type_count = Counter([cpp_function.return_type.category for cpp_function in cpp_functions if len(cpp_function.docstring) > 0])
print('Return type distribution of these functions:', type_count)
```
